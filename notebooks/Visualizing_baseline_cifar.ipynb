```lua
tnt = require 'torchnet'
image = require 'image'
Plot = require 'itorch.Plot'
require 'cunn'
require 'cudnn'
require 'gnuplot'
```

```lua
training_errors = torch.load('../logs/trainingErrors_baseline_cifar_1480451403.log')
training_losses = torch.load('../logs/trainingLosses_baseline_cifar_1480451403.log')

validation_errors = torch.load('../logs/validationErrors_baseline_cifar_1480451403.log')
validation_losses = torch.load('../logs/validationLosses_baseline_cifar_1480451403.log')

time = torch.load('../logs/timersbaseline_cifar_1480451403.log')

model = torch.load('../logs/model_baseline_cifar_1480451403.model')
```

```lua
test_accuracy = 0.94885 --After uploading on Kaggle
test_error = (1 - test_accuracy) * 100
```

```lua
print(training_errors:size())
print(validation_errors:size())
```

```lua
training_error_vector = training_errors[{{}, 276}]
validation_error_vector = validation_errors[{{}, 31}]
```

```lua
x_axis = torch.LongTensor(300)
for i=1, x_axis:size()[1] do
    x_axis[i] = i
end

test_error_vector = torch.DoubleTensor(300):fill(test_error)
```

```lua
Plot()
:line(x_axis, training_error_vector,'red','training error')
:line(x_axis, validation_error_vector,'green','validation error')
:line(x_axis, test_error_vector,'blue','validation error')
:xaxis('Epochs'):yaxis('Percentage Error')
:legend(true):title('comparing errors'):draw()
```

## Observations

1. Training error flattens out to zero quite early (30-40 epochs)
2. Validation error moves with the training error but never hits zero.
3. Test Error after the full training is high comapred to validation/training error

## Conclusions

1. Clear case of overfitting (training error is zero but validation and test error is comparatively high). Better regularization is needed.
2. Not more than 30-50 epochs is needed to fully train the model.
3. Validation set is not a good approximation of the test set. (More on this later)
